```python
phq9 = pd.read_csv(path, header = 0)
phq9.head()
```

# Task 1 : Convert Text Responses to Scores

```python
#Establish scoring map using PHQ9
scoring_map = {
    "Not at all":0,
    "Several days":1,
    "More than half the days":2,
    "Nearly every day":3
}

# Convert text responses to scores
for column in phq9.columns[2:]:  # Skip the 'id' column
    phq9[column] = phq9[column].map(scoring_map)

phq9.head()



```

```python
#creating a total score column

phq9['Total Score'] = phq9.iloc[:, 2:10].sum(axis=1)
phq9.head()
```

```python
#Create a new csv file 
phq9.to_csv('/Users/kimberlybrooks/Desktop/PythonSemester2 ', index= False)
```

# Task 2: Calculate Average PHQ-9 Scores (Pre and Post)


```python
#separating data into pre and post 

type_group = phq9.groupby("type")
pre_survey = type_group.get_group("pre")

#pre-survey mean 

pre_average = pre_survey["Total Score"].mean()
print("Pre Average :", pre_average)
```

```python
#separating data into pre and post 

type_group = phq9.groupby("type")
post_survey = type_group.get_group("post")

#pre-survey mean 

post_average = post_survey["Total Score"].mean()
print("Post Average :", post_average)
```

# Task 3 : Identify Participants with Increased Depression Scores


```python
#identify participants whose PHQ-9 scores increased from pre to post

pivot_phq9 = phq9.pivot(index='uid', columns='type', values='Total Score')

#filter for patients where post survey score is greater than their pre survey score 
increased_scores = pivot_phq9[pivot_phq9['post'] > pivot_phq9['pre']]

increased_scores = increased_scores[['pre', 'post']]

increased_scores
```
